```python
import pandas
```

```python
dataframe = pandas.read_csv('diabetes.csv')
dataframe.head()
```

```python
dataframe.info()
```

```python
attributes = dataframe[dataframe.columns[:8]]
labels = dataframe['Outcome']
```

```python
# standarization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(attributes)
attributes = scaler.transform(attributes)
```

```python
# split into training and testing data
from sklearn.model_selection import train_test_split

splitted_data = train_test_split(attributes, labels, test_size=0.33, random_state=42)
attribute_trains, attribute_tests, label_trains, label_tests = splitted_data
```

```python
# melatih model SVM (Support Vector Machine) Classifier
from sklearn.svm import SVC

classifier = SVC()
classifier.fit(attribute_trains, label_trains)

score = classifier.score(attribute_tests, label_tests) * 100
print(f'model score (accuracy): {score:.1f}%')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'poly', 'sigmoid']
}


grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3, cv=5)
grid.fit(attribute_trains, label_trains)

best_model = grid.best_estimator_
score = best_model.score(attribute_tests, label_tests) * 100

print(f'Model score (accuracy) with best parameters: {score:.1f}%')
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(best_model, attributes, labels, cv=10)
print(f"Cross-validation scores: {scores}")
print(f"Mean cross-validation score: {scores.mean() * 100:.1f}%")
```
